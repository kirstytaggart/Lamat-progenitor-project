```python
#update the new dataset with new headers with filter for dolphot to read
path = r"C:\Users\domdo\Downloads\MAST_2023-06-13T19_46_45.905Z\HST"
dir_list = os.listdir(path)
#for each file in path
for file in os.listdir(path):
    filename = os.fsdecode(file)
    #if file ends with _drz.fits, print filename, date, instrument, exp time, filter
    if filename.endswith(".fits"):
            image_path = path +"/" + os.path.basename(filename)
            print(filename)
            image = fits.open(image_path, mode = "update")
            header = image[0].header
            filter = ""
            #different fits have different filter name, sort through common 3
            if "FILTER1" in header:
                if header["FILTER1"] == "F814W" or header["FILTER2"] == "F814W":
                    filter = "F814W"
                elif header["FILTER1"] == "F555W" or header["FILTER2"] == "F555W":
                    filter = "F555W"
                else:
                    filter = "NOT DETECTED"
                header.set("FILTER", filter, before="FILTER1")
                image.flush()
                print(header["DATE-OBS"] + "\n" + header["INSTRUME"] + "\n" + str(header["EXPTIME"]) + "\n" + header["FILTER"])
            image.close()
```
